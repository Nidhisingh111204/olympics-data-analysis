```python
def fetch_medal_tally(df, year, country):
    medal_df=df.drop_duplicates(subset=['Team',	'NOC',	'Games',	'Year',	'Season',	'City',	'Sport',	'Event',	'Medal'	])
    flag=0
    if year == 'Overall' and country=='Overall':
        temp_df=medal_df
    if year == 'Overall' and country!='Overall':
        flag=1
        temp_df=medal_df[medal_df['region']==country]
    if year != 'Overall' and country=='Overall':
        temp_df= medal_df[medal_df['Year']==int(year)]
    if year != 'Overall' and country!='Overall':
        temp_df=medal_df[(medal_df['Year']==2016) & (medal_df['region']==country)]
    if flag==1:
        x=temp_df.groupby('Year')[['Gold','Silver','Bronze']].sum().sort_values('Year').reset_index()
    else:
        x=temp_df.groupby('region')[['Gold','Silver','Bronze']].sum().sort_values('Gold', ascending=False).reset_index()
   
    x['total']=x['Gold'] +x['Silver'] +x['Bronze']

    print(x)
```

```python
def most_successful(df, sport):
    
    temp_df = df.dropna(subset=['Medal'])
    temp_df = temp_df.drop_duplicates(subset=['Name','Games','Event','Medal'])

    if sport != 'Overall':
        temp_df = temp_df[temp_df['Sport'] == sport]

    x = temp_df.groupby('Name').count()['Medal'].sort_values(ascending=False).reset_index()
    x.columns = ['Name','Medals']
    x = x.head(15)

    x = x.merge(df[['Name','Sport','region']], on='Name', how='left')
    x = x.drop_duplicates('Name')

    return x
```

```python
def most_successful(df, country):
    
    temp_df = df.dropna(subset=['Medal'])
    temp_df = temp_df[temp_df['region'] == country]

    x = temp_df['Name'].value_counts().reset_index().head(15)

    x = x.merge(
        df,
        left_on='Name',
        right_on='Name',
        how='left'
    )[['Name', 'count', 'Sport']]

    x.rename(columns={'count': 'Medals'}, inplace=True)

    x = x.drop_duplicates('Name')

    return x
```

```python
# Count male athletes per year
men = athlete_df[athlete_df['Sex'] == 'M'].groupby('Year').count()['Name'].reset_index()

# Count female athletes per year
women = athlete_df[athlete_df['Sex'] == 'F'].groupby('Year').count()['Name'].reset_index()

# Merge both datasets
final = men.merge(women, on='Year', how='left')

# Rename columns
final.rename(columns={'Name_x': 'Male', 'Name_y': 'Female'}, inplace=True)

# Replace NaN with 0
final.fillna(0, inplace=True)

# Plot graph
import plotly.express as px

fig = px.line(final, x="Year", y=["Male", "Female"])
fig.show()
```
